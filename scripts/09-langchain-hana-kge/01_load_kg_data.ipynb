## Select some data from the loaded graph

```python
sparql_command='''
	SELECT *
	FROM <kgdocu_movies>
	WHERE {
		?s ?p ?o .
	}
	LIMIT 5
'''
```

```python
params_in = (
    sparql_command, 
    request_headers, 
    None, #return parameter 1
    None  #return parameter 2
)
```

```python
params_out = mycursor.callproc("SPARQL_EXECUTE", params_in)
```

```python
content_type = next(
    (line.split(':', 1)[1].strip()
     for line in params_out[3].split('\n')
     if line.lower().startswith('content-type:')),
    None
)

print(content_type)
```

As per https://help.sap.com/docs/hana-cloud-database/sap-hana-cloud-sap-hana-database-sparql-reference-guide/sparql-query-forms-using-sparql-execute?locale=en-US&version=LATEST the `params_out[2]` results response (for `SELECT` queries) as a string

```python
match content_type.split(';')[0].strip():
    case 'application/sparql-results+xml':
        print(params_out[2])
    case 'application/sparql-results+json':
        display(eval(params_out[2]))
    case 'text/csv':
        print(params_out[2])
    case 'text/tab-separated-values':
        print(params_out[2])
    case _:  # Default case
        print(f"Unknown content type: {content_type}")
```
